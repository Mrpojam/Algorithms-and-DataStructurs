### Decision Tree for simple decisions (bit operations)

OR operaion

<img src = "dt3.png" width=50%>

```python
def OR (a, b):
    if a == True:
        return True
    elif b == True:
        return True
    else:
        return False
```

a + b = (a'b')'

```python
def AND (a, b):
    if a == False:
        return False
    elif b == False:
        return False
    else:
        return True
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv("Social_Network_Ads.csv")
data
```

```python
y = data['Purchased']
x = data.drop('Purchased', axis=1)
```

```python
model = DecisionTreeClassifier()
model.fit(x, y)
```

#### Let's Plot Our Tree

```python
from sklearn import tree
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(model, 
                   filled=True)
fig.savefig("decistion_tree.png")
```
